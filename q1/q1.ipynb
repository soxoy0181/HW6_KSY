```python
import csv
import matplotlib.pyplot as plt

if __name__ == '__main__':
    
    region = {"전국":[], "서울":[], "대전":[], "부산":[], "제주": []}

    for reg in list(region.keys()):
        f = open(reg+'.csv','r')
        data = csv.reader(f,delimiter=',')
        header = next(data)
        
        for row in data:
            if row[2]=="": break
            region[reg].append(float(row[2]))
        
        f.close()
            
    xdata = ["{0:d}월".format(i) for i in range(1,13)]
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize=(10,10))
    plt.title("2022년 지역별 월평균 기온")
    plt.ylabel('온도(섭씨)')
    plt.xlabel('월')
    plt.plot(xdata,region['전국'], label='전국',color='red')
    plt.plot(xdata,region['서울'], label='서울',color='orange')
    plt.plot(xdata,region['대전'], label='대전',color='green')
    plt.plot(xdata,region['부산'], label='부산',color='blue')
    plt.plot(xdata,region['제주'], label='제주',color='purple')
    plt.legend()

    color = ['red','orange','green','blue','purple']
    idx = 0
    for reg in list(region.keys()):
        minVal = min(region[reg])
        maxVal = max(region[reg])
        for i in range(0,12):
            plt.text(i,region[reg][i],"{0:.1f}".format(region[reg][i]),color=color[idx])
        idx+=1
    
    plt.show()
    
    #전국평균대비 각 지역별 기온의 차이
    print("[전국 평균 대비 각 지역별 기온의 차이]")
    for i in range(0,13):
        if i==0:
            print("     ",end="")
        else:
            print("{0:4d}월".format(i),end="")
    print()
    data_all = region['전국']
    for reg in list(region.keys()):
        print(reg,end=" ")
        data = region[reg]
        for i in range(0,12):
            print("{0:6.1f}".format(data[i]),end="")
        print()
        if reg=='전국': continue
        print("      ",end="")
        for i in range(0,12):
            diff = data[i]-data_all[i]
            print("({0:+4.1f})".format(diff),end="")
        print()
    
    print("전국 대비 모든 지역은 대부분 더 높은 기온을 기록하고 있습니다.")
    print("전국 대비 가장 더운 지역은 제주시입니다.")
    print("제주시는 최고 평균 기온을 섭씨 29.2도로 기록하며 전국 최고 평균 기온인 25.9도보다 3.3도 높은 온도를 기록하였습니다.")
    print("전국대비 가장 추운 지역은 서울시입니다.")
    print("서울시는 최저 평균 기온을 섭씨 -2.8도로 기록하며 전국 최고 평균 기온인 -1.4도보다 1.4도 낮은 온도를 기록하였습니다.")
```
