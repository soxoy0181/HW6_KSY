```python
import csv
import matplotlib.pyplot as plt

if __name__ == '__main__':
    
    region = {"":[], "":[], "":[], "":[], "": []}

    for reg in list(region.keys()):
        f = open(reg+'.csv','r')
        data = csv.reader(f,delimiter=',')
        header = next(data)
        
        for row in data:
            if row[2]=="": break
            region[reg].append(float(row[2]))
        
        f.close()
            
    xdata = ["{0:d}".format(i) for i in range(1,13)]
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize=(10,10))
    plt.ylabel('()')
    plt.xlabel('')
    plt.plot(xdata,region[''], label='',color='red')
    plt.plot(xdata,region[''], label='',color='orange')
    plt.plot(xdata,region[''], label='',color='green')
    plt.plot(xdata,region[''], label='',color='blue')
    plt.plot(xdata,region[''], label='',color='purple')
    plt.legend()

    color = ['red','orange','green','blue','purple']
    idx = 0
    # 
    for reg in list(region.keys()):
        minVal = min(region[reg])
        maxVal = max(region[reg])
        for i in range(0,12):
            plt.text(i,region[reg][i],"{0:.1f}".format(region[reg][i]),color=color[idx])
#         plt.text(region[reg].index(minVal),minVal, "{0:s}  {1:.1f}".format(reg,minVal))
#         plt.text(region[reg].index(maxVal),maxVal, "{0:s}  {1:.1f}".format(reg,maxVal))
        idx+=1
    
    plt.show()
    
    #    
    print("[      ]")
    for i in range(0,13):
        if i==0:
            print("     ",end="")
        else:
            print("{0:4d}".format(i),end="")
    print()
    data_all = region['']
    for reg in list(region.keys()):
        print(reg,end=" ")
        data = region[reg]
        for i in range(0,12):
            print("{0:6.1f}".format(data[i]),end="")
        print()
        if reg=='': continue
        print("      ",end="")
        for i in range(0,12):
            diff = data[i]-data_all[i]
            print("({0:+4.1f})".format(diff),end="")
        print()
    
    print("         .")
    print("     .")
    print("     29.2      25.9 3.3   .")
    print("    .")
    print("     -2.8      -1.4 1.4   .")
```
