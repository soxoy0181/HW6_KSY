```python
import matplotlib.pyplot as plt
import csv

if __name__ == '__main__':
    f = open('2023년 03월 교통카드 통계자료.csv','r', encoding='UTF8')
    data = csv.reader(f,delimiter=',')
    header = next(data)
    header = next(data)
    
    onStation = {}
    offStation = {}
    onOffStation = {}
    
    for row in data:
        #row[7] ~ row[10]
        for i in range(7,11):
            row[i] = row[i].replace(",","")
            row[i] = int(row[i])
        onStation[row[3]] = row[7]+row[9]
        offStation[row[3]] = row[8]+row[10]
        onOffStation[row[3]] = onStation[row[3]] + offStation[row[3]]
     
    f.close()
        
    #graph
    plt.rc('font', family='Malgun Gothic')
    fig = plt.figure(figsize=(50,30))
    #승차    
    sortedOnStation = dict(sorted(onStation.items(), key=lambda x:x[1],reverse=True))
    key_sortedOnStation = list(sortedOnStation.keys())[0:30]
    val_sortedOnStation = list(sortedOnStation.values())[0:30]
    
    p1 = fig.add_subplot(3,1,1)
    p1.bar(key_sortedOnStation, val_sortedOnStation)
    p1.set_xlabel("역 이름")
    p1.set_ylabel("승차 승객수")
    p1.set_title("출근시간 최대 승차역 30개")
    
    #하차
    sortedOffStation = dict(sorted(offStation.items(), key=lambda x:x[1],reverse=True))
    key_sortedOffStation = list(sortedOffStation.keys())[0:30]
    val_sortedOffStation = list(sortedOffStation.values())[0:30]
    
    p2 = fig.add_subplot(3,1,2)
    p2.set_xlabel("역 이름")
    p2.set_ylabel("하차 승객수")   
    p2.bar(key_sortedOnStation, val_sortedOnStation)
    p2.set_title("출근시간 최대 하차역 30개")
    
    #승하차
    sortedOnOffStation = dict(sorted(onOffStation.items(), key=lambda x:x[1],reverse=True))
    key_sortedOnOffStation = list(sortedOnOffStation.keys())[0:30]
    val_sortedIbOffStation = list(sortedOnOffStation.values())[0:30]

    p3 = fig.add_subplot(3,1,3)
    p3.bar(key_sortedOnStation, val_sortedOnStation)
    p3.set_xlabel("역 이름")
    p3.set_ylabel("승하차 승객수")
    p3.set_title("출근시간 최대 승하차역 30개")
    
    plt.show()    
    
    
    print("[2023년 3월 오전 7-9시 출근시간 최대 승하차 역 분석 결과]")
    print("동일한 30개의 지하철역이 같은 순서로 출근시간 최대 승차역, 최대 하차역, 최대 승하차역 30개에 나타나고 있습니다.")
    print("출근 시간대 사람들이 가장 많이 타고 내리는 역은 강남역입니다.")
```
